```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#A Dataset is taken,EDA(Exploratory Data Analysis) is performed,
#a suitable Classifier,Regressor or Clusterer is applied
# and the accuracy of the model is calculated.
```

```python
#1.Take data to create a dataframe
df=pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/areavsprices.csv')
df
```

```python
#Here, Area[input(x)] is in sq feet and price[output(y)]in rupees

```

```python
#2.Preprocessing not required
```

```python
#3.Visualisation
plt.scatter(df['Area'],df['Prices'])
plt.title('Area vs Prices')
plt.xlabel('X-axis')
plt.ylabel("Y-axis")
plt.grid()
```

```python
#Input-Area,Output-Prices
#4.Divide data into input and output
```

```python
x=df.iloc[0:6,0:1].values
x
```

```python
y=df.iloc[:,1].values
y
```

```python
#5.Train and test Variants>Not applicable
```

```python
#6.Scaling>Not applicable
```

```python
#7.Run a classifier
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#8.Fit the model
model.fit(x,y)
```

```python
#9.Predict Output
y_Pred=model.predict(x)
y_Pred
```

```python
y
```

```python
model.predict([[2000]])
```

```python
m=model.coef_
m
```

```python
C=model.intercept_
C
```

```python
#to find y
m*2000+C
```

```python
#Visualisation for best fit line
plt.scatter(x,y)
plt.plot(x,y_Pred,c='Orange')
plt.title('Best fit line')
plt.xlabel("Area")
plt.ylabel("Prices")
```

```python
#Hence,a Dataset is taken,EDA(Exploratory Data Analysis) is performed,
#a suitable Classifier,Regressor or Clusterer is applied
# and the accuracy of the model is calculated.
#Since y is equal to model.predict,we can say that the model given above is accurate.
```
